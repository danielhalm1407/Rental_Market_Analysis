### Unpack Key Scores for Each University



#### Filter Out Only the Desired Columns

```python
def filter_df(df=results):
    results_df = pd.json_normalize(df)
    results_df.columns = results_df.columns.str.lower().str.replace('.', '_').str.replace(' ', '_')
    base_cols = [
        'score_nid',
        'nid',
        'core_id',
        'title',
        'path',
        'region',
        'country',
        'city',
        'overall_score',
        'rank',
    ]
    score_cols = results_df.filter(regex='scores').columns.tolist()
    columns_of_interest = base_cols + score_cols
    filtered_df = results_df[columns_of_interest]
    return filtered_df


filtered_df = filter_df(results)
filtered_df.head()
```

#### Create A Dataframe that Unpacks the Nested JSON into a Dataframe Containing Key Scores for Each University

```python
def explode_and_pivot_scores(filtered_df):
    adjusted_df = filtered_df.copy()
    score_cols = filtered_df.filter(regex='scores').columns.tolist()
    for col in score_cols:
        # Explode that column containing lists of dictionaries
        df_expanded = adjusted_df.explode(col)
        # Normalize the exploded column
        normalized = pd.json_normalize(df_expanded[col]).add_suffix(f"_{col}")
        # Join the normalized DataFrame back to the exploded DataFrame
        df_long = df_expanded.drop(columns=col).reset_index().join(normalized).drop(columns='index')
        # Change into wide data
        cols_to_keep = list(df_expanded.columns[~df_expanded.columns.str.contains(f'{col}')])
        df_wide = df_long.pivot_table(index='core_id', columns=f"indicator_name_{col}", values=f'score_{col}', aggfunc='first').reset_index()
        # Ensure 'core_id' is the same dtype in both DataFrames before joining
        adjusted_df['core_id'] = adjusted_df['core_id'].astype(str)
        df_wide['core_id'] = df_wide['core_id'].astype(str)
        adjusted_df = adjusted_df[cols_to_keep].merge(df_wide, on='core_id', how='left')
    
    # Nromalize the column names to be lowercase and replace spaces with underscores
    adjusted_df.columns = adjusted_df.columns.str.lower().str.replace(' ', '_')
    return adjusted_df




explode_and_pivot_scores(filtered_df).head(4)
```
